# Contemos la cantidad de tokens de un texto

```python
import tiktoken

encoding_name = tiktoken.encoding_name_for_model("text-embedding-ada-002")

encoding = tiktoken.get_encoding(encoding_name)
```

```python
encoded_text = encoding.encode("I love pets")

encoded_text
```

# Excedamos la mÃ¡xima cantidad de tokens

```python
long_text = "I love pets" * int(8193/3)
```

```python
long_encoded_text = encoding.encode(long_text)

len(long_encoded_text)
```
